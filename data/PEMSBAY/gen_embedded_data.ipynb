```python
npz_data=np.load("./PEMSBAY.npz")

npz_data["data"].shape
```

```python
index_list=df_data.index
index_list
```

```python
def encode_timestamp(ts):
    weekday_embedding=np.eye(7)[ts.weekday()].tolist()
    hour_embedding=np.eye(24)[ts.hour].tolist()
    holiday_embedding=[1.0 if ts.weekday()>4 else 0.0]
    
    embedding=weekday_embedding+hour_embedding+holiday_embedding
    return embedding

embeddings=[]
for ts in index_list:
    embeddings.append(encode_timestamp(ts))
    
embeddings=np.array(embeddings)
embeddings.shape
```

```python
time_embeddings=np.broadcast_to(embeddings[:, np.newaxis, :], (embeddings.shape[0], df_data.shape[1], embeddings.shape[1]))

time_embeddings.shape
```

```python
df_se=pd.read_csv("./SE_pemsbay.txt", skiprows=1, index_col=[0], sep="\s+", header=None)

df_se
```

```python
df_se.sort_index(inplace=True)

df_se
```

```python
spatial_embeddings=df_se.values

spatial_embeddings.shape
```

```python
spatial_embeddings=np.broadcast_to(spatial_embeddings[np.newaxis, :, :], (time_embeddings.shape[0], spatial_embeddings.shape[0], spatial_embeddings.shape[1]))

spatial_embeddings.shape
```

```python
data_embedded=np.concatenate((df_data.values[..., np.newaxis], spatial_embeddings, time_embeddings), axis=2)

data_embedded.shape
```

```python
np.save("./PEMSBAY_embedded.npy", data_embedded)
```

```python
np.savez_compressed("./PEMSBAY_embedded.npz", data=data_embedded)
```
